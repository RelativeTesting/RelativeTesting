```python
import ast
import astor

code_file="code_1.py"
with open(code_file, 'r') as f:
        code = f.read()

tree = ast.parse(code)

for node in ast.walk(tree):
    if isinstance(node, ast.BinOp):
        if isinstance(node.left, ast.Name):
            left = node.left.id
        elif isinstance(node.left, ast.Num):
            left = node.left.n
        elif isinstance(node.left, ast.BinOp):
            left = astor.to_source(node.left).strip()
        if isinstance(node.right, ast.Name):
            right = node.right.id
        elif isinstance(node.right, ast.Num):
            right = node.right.n
        elif isinstance(node.right, ast.BinOp):
            right = astor.to_source(node.right).strip()
        operator = type(node.op).__name__
        print(f"{left} {operator} {right}")
```
